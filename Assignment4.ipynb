### Glossary

https://www.nba.com/stats/help/glossary/

```python
# 3PA
# Name: 3 Point Field Goals Attempted
# Definition: The number of 3 point field goals that a player or team has attempted

# 3PM
# Name: 3 Point Field Goals Made
# Definition: The number of 3 point field goals that a player or team has made

# 3P%
# Name: Percent of Team's 3 Point Field Goals Made
# Definition: The percentage of a team's 3 point field goals that a player has made while on the court

# PTS
# Name Points
# Definition The number of points scored
```

```python
df = df_1997_2022[['PTS', '3PM', '3PA']]

df.head()
```

```python
# Reverse table
df = df.reindex(index=df.index[::-1])

df.head()
```

```python
df.tail()
```

```python
df['PTS'].describe()
```

```python
option = '3PA'

print('{}: {:.2f}'.format(option, df[option]['2021-22'] / df[option]['1996-97']))
```

```python
option = '3PM'

print('{}: {:.2f}'.format(option, df[option]['2021-22'] / df[option]['1996-97']))
```

```python
option = 'PTS'

print('{}: {:.2f}'.format(option, df[option]['2021-22'] / df[option]['1996-97']))
```
